```python
from importlib.resources import open_binary
from cchdo.hydro import read_csv
from cchdo.hydro import accessors

with open_binary("cchdo.hydro.tests.data", "merge_placeholders.csv") as f:
    data = read_csv(f)
data.oxygen
```

```python
str_fq = [
  {
    "EXPOCODE": "TEST",
    "STNNBR": "1",
    "CASTNO": 1,
    "SAMPNO": "36",
    "OXYGEN [UMOL/KG]": "234.11",
    "OXYGEN [UMOL/KG]_FLAG_W": "2",
  }
]
literal_fq = [
  {
    "EXPOCODE": "TEST",
    "STNNBR": "1",
    "CASTNO": 1,
    "SAMPNO": "36",
    "OXYGEN [UMOL/KG]": 234.11,
    "OXYGEN [UMOL/KG]_FLAG_W": "2",
  }
]
```

```python
merged_str = data.cchdo.merge_fq(str_fq)
merged_str.oxygen
```

```python
merged_lit = data.cchdo.merge_fq(literal_fq)
merged_lit.oxygen
```

```python
print(merged_str.cchdo.to_exchange().decode())
print(merged_lit.cchdo.to_exchange().decode())
```

```python
import polars as pl
import requests
from io import BytesIO
```

```python
# The thing to notice here is that polars read the underlying data type/property and ignored the formatting within excel
# You (the merger) want to keep both of these and will need to manually set the C_format
# TODO bookkeping attr updater in hydro?
file = BytesIO(requests.get("https://cchdo.ucsd.edu/data/44600/data.xlsx").content)
in_data = pl.read_excel(file, sheet_name="DIC")
in_data
```

```python
literal_fq = [
  {
    "EXPOCODE": "TEST",
    "STNNBR": "1",
    "CASTNO": 1,
    "SAMPNO": "36",
    "OXYGEN [UMOL/KG]": None, # just ignore that this is DIC data going into the wrong param
    "OXYGEN [UMOL/KG]_FLAG_W": "5",
  }
]
literal_fq
```

```python
merged_lit = data.cchdo.merge_fq(literal_fq)
merged_lit.oxygen
```

```python
#hmm... 2177.0 is not the same as 2176.99 that I see with my eyes in excel
print(merged_lit.cchdo.to_exchange().decode())
```

```python
# update the C format so text formats print the same as formatted in excel
# This can be used to fix "print precision" mistakes, _only_ the print attribute needs to be changed, this will even work after writing to netCDF and read back in.
merged_lit.oxygen.attrs["C_format"] = "%.2f"
print(merged_lit.cchdo.to_exchange().decode())
```
